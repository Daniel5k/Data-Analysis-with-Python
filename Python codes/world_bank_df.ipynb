```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

world_bank_df = pd.read_csv(r'C:\Users\DANIEL\Documents\VS CODE FILES\Data Analyst Projects\Kaggle datasets\Datasets real files\world_bank_dataset.csv')

filter_for_highest_life_expectancy = world_bank_df.groupby('Country')['Life Expectancy'].agg([np.mean, np.median, np.std,])

print(filter_for_highest_life_expectancy.sort_values(by='mean', ascending=True))
# filter_for_highest_life_expectancy.hist()
filter_for_highest_life_expectancy.plot(kind='bar')
plt.show()


''' Brazil and Turkey has the highest life expectancy rate'''
```

```python
# <==== List of Countries filtered by Co2 emission ====>
countries_with_the_highest_emissonOfC02 = world_bank_df.groupby('Country')['CO2 Emissions (metric tons per capita)'].agg(sum).sort_values(ascending=False)
countries_with_the_highest_emissonOfC02.plot(kind='bar')
countries_with_the_highest_emissonOfC02.reset_index()

'''from the statistics India and france has the highest number of CO2 emission 
and this does not affect their GDP
'''

```

```python
country_with_the_most_progress  = world_bank_df.groupby('Country')['GDP (USD)'].agg(sum).sort_values(ascending=False)
country_with_the_most_progress

#concatinating the values of progress of countries and their carbon emission
df = pd.concat([country_with_the_most_progress,countries_with_the_highest_emissonOfC02], axis=1)
df.sort_values(by='CO2 Emissions (metric tons per capita)',ascending=False)



#plotting the values of CO2 emission and GDP of different countries
df[['CO2 Emissions (metric tons per capita)']].plot(kind='bar', color='red')
df[['GDP (USD)']].plot(kind='bar')
plt.show()


'''
Italy and Saudi Arabia has the highest progress
'''
```

```python
world_bank_df['log_GDP (USD)'] = np.log(world_bank_df['GDP (USD)'])

sns.lmplot(x='log_GDP (USD)', y='CO2 Emissions (metric tons per capita)', data = world_bank_df, ci=None)
plt.show()
world_bank_df['log_GDP (USD)'].corr(world_bank_df['CO2 Emissions (metric tons per capita)'])
```
