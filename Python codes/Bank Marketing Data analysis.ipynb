**AGE DISTRIBUTION FOR CUSTOMERS WITH LOAN**

```python
# import matplotlib.pyplot as plt

df_age_with_loan = df[df['loan'] == 'yes']



different_ages_in_df = df_age_with_loan['age']

age_distribution_of_customers_with_loan = {
    '0-17': 0,
    '18-35': 0,
    '36-50': 0,
    '51-100': 0,
}

for age in different_ages_in_df:
    if age <=17:
        age_distribution_of_customers_with_loan['0-17'] +=1
    elif age <=35:
        age_distribution_of_customers_with_loan['18-35'] +=1
    elif age <=50:
        age_distribution_of_customers_with_loan['36-50'] +=1
    else:
        age_distribution_of_customers_with_loan['51-100'] +=1

total_of_people_with_loan = 0
distribution_with_loan_values = []
for key, value in age_distribution_of_customers_with_loan.items():
    total_of_people_with_loan += value
    distribution_with_loan_values.append(value)

plt.hist(x=distribution_with_loan_values, color=['orange'])
plt.title('distribution with loan values')
plt.show()


print(f'The total number of people that took loan is : {total_of_people_with_loan}')

age_distribution_of_customers_with_loan
```

**AGE DISTRIBUTION FOR ALL CUSTOMERS**

```python
df_age = df['age']

age_distribution_of_all_customers = {
    '0-17': 0,
    '18-35': 0,
    '36-50': 0,
    '51-100': 0,
}

for age in df_age:
    if age <=17:
        age_distribution_of_all_customers['0-17'] +=1
    elif age <=35:
        age_distribution_of_all_customers['18-35'] +=1
    elif age <=50:
        age_distribution_of_all_customers['36-50'] +=1
    else:
        age_distribution_of_all_customers['51-100'] +=1

age_distribution_of_all_customers
age_dataframe = pd.DataFrame([age_distribution_of_all_customers])
age_dataframe = age_dataframe.melt()
sns.set_style('whitegrid')

sns.catplot(data=age_dataframe, y='value',x='variable', hue='variable', kind='bar')
plt.show()
age_dataframe
```
